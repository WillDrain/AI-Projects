```python
# Iris Dataset Cleaning - Day 3
import pandas as pd
import numpy as np
```

```python
# Load Dataset
df = pd.read_csv("../data/iris.csv")
df.head()
```

```python
# Inspect dataset
df.info()
df.describe(include="all")
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Rename columns
df = df.rename(columns={
    "SepalLengthCm": "sepal_length",
    "SepalWidthCm": "sepal_width",
    "PetalLengthCm": "petal_length",
    "PetalWidthCm": "petal_width",
    "Species": "species"
})
df.head()
```

```python
# Drop unnecessary columns
if "Id" in df.columns:
    df = df.drop(columns=["Id"])
df.head()
```

```python
# Save Cleaned dataset
df.to_csv("../data/iris_cleaned.csv", index=False)
print("Cleaned dataset saved to iris_cleaned.csv")
```

```python
# Vector Norm
v = np.array([3, 4])
norm = np.linalg.norm(v)
print("Vector:", v)
print("Norm:", norm)
```

```python
#Normalization
normalized = v / norm
print("Normalized vector:", normalized)
```
